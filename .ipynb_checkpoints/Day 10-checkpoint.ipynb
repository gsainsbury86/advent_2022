```python
import requests
year = 2022
day = 9
url = f"https://adventofcode.com/{year}/day/{day}/input"

with open('cookie.txt', 'r') as file:
    session = file.read()

headers = {'cookie': f"session={session}"}

with open(f'day_{day}_input.txt', 'wb') as file:
    file.write(requests.get(url, headers=headers).content)
```

```python
import math

def move(head_pos, last_tail_pos, dim):
    if head_pos[dim] > last_tail_pos[dim]:
        new_pos[dim] = last_tail_pos[dim] + 1
    else:
        new_pos[dim] = last_tail_pos[dim] - 1
    return new_pos

head_pos = [0,0]
tail_pos = [(0,0)]
for line in lines:
    dirn, s_distance = line.split(" ")
    dist = int(s_distance)
    for i in range(0,dist):
        if dirn == 'R':
            head_pos[0] = head_pos[0] + 1
        if dirn == 'L':
            head_pos[0] = head_pos[0] - 1
        if dirn == 'U':
            head_pos[1] = head_pos[1] + 1
        if dirn == 'D':
            head_pos[1] = head_pos[1] - 1
    
        last_tail = list(tail_pos[-1])
        lag = math.hypot(last_tail[0] - head_pos[0], last_tail[1] - head_pos[1])
        #print(lag)
        if lag > 1.5:
            # Move Required?
            new_pos = last_tail
            for d in [0,1]:
                if abs(head_pos[d] - last_tail[d]) > 0:
                    new_pos = move(head_pos, new_pos, d)
            tail_pos.append(tuple(new_pos))
        #print(head_pos, tail_pos[-1])  
                
            
print(len(set(tail_pos)))
```

```python
import math

def move(head_pos, last_tail_pos, dim):
    if head_pos[dim] > last_tail_pos[dim]:
        new_pos[dim] = last_tail_pos[dim] + 1
    else:
        new_pos[dim] = last_tail_pos[dim] - 1
    return new_pos

#head_pos = [0,0]
tail_pos = [(0,0)]

knot_position_history = [[(0,0)] for x in range(0,10)]

print(knot_position_history)

for line in lines:
    dirn, s_distance = line.split(" ")
    dist = int(s_distance)
    lead_knot = list(knot_position_history[0][-1])
    for i in range(0,dist):
        if dirn == 'R':
            lead_knot[0] = lead_knot[0] + 1
        if dirn == 'L':
            lead_knot[0] = lead_knot[0] - 1
        if dirn == 'U':
            lead_knot[1] = lead_knot[1] + 1
        if dirn == 'D':
            lead_knot[1] = lead_knot[1] - 1
        knot_position_history[0].append(tuple(lead_knot))
        
        for knot_i, knot in enumerate(knot_position_history[1:]):
            last_tail = list(knot[-1])
            head_pos = knot_position_history[knot_i][-1]
            #print(last_tail,head_pos)
            lag = math.hypot(last_tail[0] - head_pos[0], last_tail[1] - head_pos[1])
            #print(lag)
            #print(lag)
            if lag > 1.5:
                # Move Required?
                new_pos = last_tail
                for d in [0,1]:
                    if abs(head_pos[d] - last_tail[d]) > 0:
                        new_pos = move(head_pos, new_pos, d)
                knot_position_history[knot_i+1].append(tuple(new_pos))
            #print(head_pos, tail_pos[-1])  
                
print(knot_position_history)            
print(len(set(knot_position_history[-1])))
```
