```python
import requests
year = 2022
day = 24

url = f"https://adventofcode.com/{year}/day/{day}/input"

with open('cookie.txt', 'r') as file:
    session = file.read()

headers = {'cookie': f"session={session}"}

with open(f'inputs/day_{day}_input.txt', 'wb') as file:
    file.write(requests.get(url, headers=headers).content)
    
import os
if os.path.exists(f"inputs/day_{day}_test.txt"):
    test = [x for x in open(f'inputs/day_{day}_test.txt', 'r')]
    for line in test[:10]:
        print(line)
print("---")
if os.path.exists(f"inputs/day_{day}_test_2.txt"):
    test_2 = [x for x in open(f'inputs/day_{day}_test_2.txt', 'r')]
    for line in test_2[:10]:
        print(line)
print("---")
lines = [x for x in open(f'inputs/day_{day}_input.txt', 'r')]
for line in lines[:10]:
    print(line)
```

```python
from copy import deepcopy
run_test = True

if run_test:
    case = test_2
else:
    case = lines

case = [x.strip() for x in case]
    
# E = []
B = []
W = []

# UP DOWN LEFT RIGHT
direction_arrows = ['^','v','<','>']
directions = [(-1,0),(1,0),(0,-1),(0,1)]

for d in direction_arrows:
    B.append([])
    
for r, R in enumerate(case):
    for c, C in enumerate(R):
        # if C == '.':
        #     E.append((r,c))
        if C == '#':
            W.append((r,c))
        elif C in direction_arrows:
            B[direction_arrows.index(C)].append((r,c))

for i, b in enumerate(B):
    B[i] = tuple(b)            
            
def show(B,W,p):
    for r, R in enumerate(case):
        output = ""
        for c, C in enumerate(R):
            if p == (r,c):
                output = output + 'E'
            else:
                blizz_count = []
                for i, b in enumerate(B):
                    if (r,c) in b:
                        blizz_count.append(direction_arrows[i])
                if len(blizz_count) == 1:
                        output = output + blizz_count[0]
                elif len(blizz_count) == 0:
                    if (r,c) in W:
                        output = output + '#'
                    else:
                        output = output + '.'
                elif len(blizz_count) <= 10:
                    output= output + str(len(blizz_count))
                else:
                    output = output + '@'

        print(output)
            
show(B,W,(0,1))

# E_s = E.copy()
B_s = deepcopy(B)
W_s = W.copy()


```

```python
from collections import deque

B = deepcopy(B_s)
W = W_s.copy()
# E = E_s.copy()


def update_blizzards(B):
    B = deepcopy(B)
    B = list(map(list, B))
    
    for d_i, d_a in enumerate(direction_arrows):
        for b_i, b in enumerate(B[d_i]):
            d = directions[d_i]
            new_b = [b[0] + d[0], b[1] + d[1]]

            #if new_b in W:
            if new_b[0] == 0 or new_b[0] == ROWS - 1:
                if d_a == direction_arrows[0]:
                    new_b[0] = ROWS - 2
                if d_a == direction_arrows[1]:                    
                    new_b[0] = 1
            elif new_b[1] == 0 or new_b[1] == COLS - 1:
                if d_a == direction_arrows[2]:                    
                    new_b[1] = COLS - 2
                if d_a == direction_arrows[3]:              
                    new_b[1] = 1
            #print(new_b)
            B[d_i][b_i] = tuple(new_b)
            
    B = tuple(map(tuple, B))

    return B


# B = update_blizzards(B)
# show(B,W,(ROWS - 1,COLS -2))

# assert False

                
ROWS = len(case)
COLS = len(case[0])

p_0 = (0,1)

Q = deque([(p_0, tuple(B), 0)])

states = set()

while Q:
    state = Q.popleft()
    #print(state)
    
    if state in states:
        continue
    states.add(state)
    
    if len(states) % 5000 == 0:
        print(len(states))
    
    p = state[0]
    B = state[1]
    time = state[2]
    

    if p[0] == ROWS - 1 and p[1] == COLS -2:
        print(f"Completed at time: {time}")
        #show(B,W,p)
        break
    
    if p in W:
        continue
        
        
    if p[0] <= 0 or p[0] >= ROWS - 1 or p[1] <= 0 or p[1] >= COLS - 1:
        #print("OOB")
        continue

    for d_i, d_a in enumerate(direction_arrows):
        if p in B[d_i]:
            continue
            
    B = update_blizzards(B)
    for d in directions + [(0,0)]:
        Q.append(((p[0]+d[0],p[1]+d[1]), B, time+1))
```
